```python
#iport para inciso 6 al 9
import numpy as np
import pandas as pd
import re
import nltk 
from collections import Counter
from nltk.corpus import stopwords
from PIL import Image
import matplotlib.pyplot as plt
from nltk.util import ngrams
from nltk import word_tokenize
from nltk.corpus import reuters
from collections import Counter, defaultdict
import random
```

```python
#5% del texto
twitter = open("C:/Users/Estuardo/Downloads/6261_9186_compressed_final_en_US_en_US.twitter.txt/en_US.tw.txt", encoding="utf8").read()
tweets = twitter.lower()
```

```python
#matrices de 2-gramas

from nltk import ngrams
n = 2
dosgrams = ngrams(tweets.split(), n)

for grams2 in dosgrams:
  print(grams2)
```

```python
#matriz de 3-gramas
three_gram=[]
size=3

tresgrams = ngrams(tweets.split(), size)

for grams3 in tresgrams:
  print(grams3)
```

```python
#probabilidad de ocurrencia de cada uno de los n-gramas, puede utilizar Kneser-Ney 
#smoothing probability (KNP), para predecir las siguientes 3 palabras


from nltk.lm import KneserNeyInterpolated
from nltk.lm.preprocessing import padded_everygram_pipeline

n = 3 # Order of ngram
train_data, padded_sents = padded_everygram_pipeline(n, tweets)

model = KneserNeyInterpolated(n) 
model.fit(train_data, padded_sents)

gut_ngrams = nltk.ngrams(tweets,3)
freq_dist = nltk.FreqDist(gut_ngrams)
kneser_ney = nltk.KneserNeyProbDist(freq_dist)
print(kneser_ney)
print(freq_dist)
print(model)
```
